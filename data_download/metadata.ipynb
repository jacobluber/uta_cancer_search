```python
minimal_fields = ['file_name', 'file_size', 'state', 'md5sum', 'cases.primary_site']

fields = ",".join(minimal_fields)

files_endpt = "https://api.gdc.cancer.gov/files"

filters = {
    "op": "=",
    "content": {
        "field": "data_format",
        "value": ["SVS"]
    }
}

params = {
    "filters": json.dumps(filters),
    "fields": fields,
    "format": "CSV",
    "size": "50000"
}

response = requests.post(files_endpt, headers={"Content-Type": "application/json"}, json=params)
dataframe = pd.read_csv(io.StringIO(response.content.decode('utf-8')), dtype='object')

name_dict = {
    'cases.0.primary_site': 'primary_site',
    'file_name': 'filename',
    'file_size': 'size',
    'md5sum': 'md5'
}
dataframe = dataframe.rename(name_dict, axis='columns').copy(deep=True)

cols = ['id', 'filename', 'md5', 'size', 'state', 'primary_site']
dataframe = dataframe[cols].copy(deep=True)

dataframe.to_csv('minimal_metadata.csv', index=False)
```
